```python
'''

9 - Encontre todos os carros cujo valor está entre 100 e 150.'''

(carro.query('valor>=100 & valor>=150')[['modelo','valor']].sort_values(by='valor'))

```

```python
'''14 - Agrupe os clientes por estado e conte quantos clientes existem em cada estado.'''

total = cliente[['estado','codcliente']].groupby('estado').count()
total.rename({'codcliente':'total de clientes'},axis =1, inplace=True)
display(total)
```

```python
'''
15 - Agrupe os aluguéis por cliente e filtre apenas aqueles que fizeram mais de 2 aluguéis.
'''
(pd.merge(aluguel,cliente, on='codcliente',how ='inner').groupby('nome').agg(qtd_de_alugueis = ('codaluguel','count')).query('qtd_de_alugueis >2'))
```

```python
'''
16 - Liste todos os aluguéis, mostrando o nome do cliente e o modelo do carro alugado.
'''

carros = pd.merge(carro, aluguel, on='codcarro',how='inner')
(pd.merge(carros,cliente, on='codcliente', how='inner')[['nome','modelo']].sort_values(by='nome'))
```

```python
'''
17 - Liste todos os carros e, se houver, as informações sobre os aluguéis realizados desses carros.
'''

carro.merge(aluguel, on='codcarro', how='left')[['modelo','data_aluguel']].sort_values(by='data_aluguel')
```

```python
'''

18 - Liste todos os aluguéis e as informações sobre os carros alugados, mesmo que o carro não esteja mais no banco de dados.

'''

(aluguel.merge(carro, on ='codcarro',how='left')[['modelo','data_aluguel']].sort_values(by='data_aluguel'))
```

```python
'''
19 - Liste todos os clientes e os carros que alugaram, incluindo clientes que não alugaram e carros que não foram alugados.

'''

(cliente.merge(aluguel, on='codcliente', how='outer')
.merge(carro, on='codcarro', how='outer').sort_values(by='nome')
[['nome','modelo']])
```

```python
'''
20 - Combine os resultados de duas consultas: uma que lista clientes solteiros e outra que lista clientes casados.
'''

solteiros = cliente[cliente['estadocivil'] == 'S']
casados = cliente[cliente['estadocivil'] == 'C']
pd.concat([solteiros, casados], ignore_index=True)
```

```python
'''

22 - Liste os modelos de carros e o número de vezes que cada um foi alugado.

'''

carros = pd.merge(carro, aluguel, on='codcarro', how='left')
(carros.groupby('modelo', as_index=False).agg(qtd_alugueis=('codaluguel','count'))
.sort_values(by='qtd_alugueis', ascending=False))
```
